# Business Understanding



#**Latar Belakang**


Dataset yang digunakan berisi data karakteristik morfologi bunga iris yang terdiri atas tiga spesies, yaitu Iris setosa, Iris versicolor, dan Iris virginica. Setiap spesies memiliki perbedaan ukuran pada bagian kelopak (sepal) dan mahkota (petal). Perbedaan tersebut direpresentasikan dalam empat variabel numerik, yaitu sepal length, sepal width, petal length, dan petal width.

Permasalahan dalam analisis ini adalah bagaimana memanfaatkan keempat atribut tersebut untuk mengidentifikasi spesies bunga iris secara akurat. Dengan demikian, dibutuhkan suatu metode analisis yang mampu mengelompokkan atau mengklasifikasikan spesies berdasarkan pola ukuran morfologi masing-masing bunga.

#**Tujuan Bisnis**

Tujuan dari analisis yang dilakukan meliputi:
- Mengkaji hubungan antara ukuran sepal dan petal dengan perbedaan spesies bunga iris.
- Membangun model klasifikasi yang mampu memprediksi jenis bunga iris berdasarkan data pengukuran yang tersedia.
- Mengembangkan sistem identifikasi spesies yang lebih efektif dan konsisten guna meminimalkan ketergantungan pada proses identifikasi secara manual.

#**Manfaat**

Adapun manfaat yang diharapkan dari analisis ini adalah:

- Memberikan contoh penerapan tahapan metodologi CRISP-DM dalam kasus klasifikasi data.

- Menjadi ilustrasi implementasi teknik machine learning dalam proses pengenalan pola.

- Mendukung pembelajaran mengenai pembangunan model klasifikasi berbasis data secara terstruktur dan sistematis.
